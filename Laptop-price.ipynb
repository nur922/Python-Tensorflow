```python
y=df["Price"].values
```

```python
x=df.drop(["Price","Brand"], axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test=train_test_split(x,y , test_size=0.33 , random_state=50)
```

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
x_train.shape
```

```python
x_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(18, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(9, activation="relu"))
model.add(Dense(6, activation="relu"))
model.add(Dense(3, activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam" , loss="mse")
```

```python
model.fit(x=x_train , y=y_train, validation_data=(x_test,y_test), batch_size=100 , epochs=150)
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
deney=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
mean_absolute_error(y_test,deney)
```

```python
plt.scatter(y_test,deney)
plt.plot(y_test,y_test,"r-*")
```
