```python
dataFrame2=dataFrame.drop("transmission", axis=1)
dataFrame2
```

```python
y=dataFrame2["price"].values
x=dataFrame2.drop("price",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train ,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model=Sequential()
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test) ,batch_size=300, epochs=300)
```

```python
kayipveri=pd.DataFrame(model.history.history)
```

```python
kayipveri.head(15)
```

```python
kayipveri.plot()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
testveri=model.predict(x_test)
```
