```python
dataframe=dataframe.drop(["Num_Bathrooms","Neighborhood_Quality","Num_Bedrooms"],axis=1)
```

```python
y=dataframe["House_Price"].values
x=dataframe.drop("House_Price",axis=1).values
```

```python
y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train,y_test=train_test_split(x,y , test_size=0.33 , random_state=25)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
x_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model=Sequential()
model.add(Dense(15 ,activation ="relu"))
model.add(Dense(10 ,activation ="relu"))
model.add(Dense(13,activation ="relu"))
model.add(Dense(10 ,activation ="relu"))
model.add(Dense(4 ,activation ="relu"))
model.add(Dense(1))
model.compile(optimizer="adam", loss="mse")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test,y_test), batch_size=380,epochs=500 , verbose=1 ,callbacks=[earlyStopping])
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```
