```python
df=df[df["Primary_Cam"]!=5]
df.groupby("Primary_Cam").count()
```

```python
y=df["Price"].values
```

```python
x=df.drop("Price",axis=1).values
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test,y_train ,y_test=train_test_split(x,y , test_size=0.3 , random_state=20)
```

```python
len(x_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
skaler=MinMaxScaler()
```

```python
x_train=skaler.fit_transform(x_train)
```

```python
x_test=skaler.fit_transform(x_test)
```

```python
x_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model=Sequential()
model.add(Dense(150, activation="relu"))
model.add(Dense(75, activation="relu"))
model.add(Dense(75, activation="relu"))
model.add(Dense(75, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam" , loss="mse")
```

```python
early_Stop=EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True)
```

```python
model.fit(x=x_train,y=y_train, validation_data=(x_test,y_test), batch_size=100 , epochs =300, callbacks=[early_Stop])
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
deney=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,deney)
```

```python
plt.scatter(y_test,deney)
plt.plot(y_test,y_test,"r-*")
```
