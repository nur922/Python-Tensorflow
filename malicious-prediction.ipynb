```python
y=df["Type"].values
```

```python
x=df.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train ,y_test=train_test_split(x,y , test_size=0.25 , random_state=25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
x_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(units=90,activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units=45,activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units=45,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
early_Stop=EarlyStopping(
    monitor='val_loss',
    patience=100,
    verbose=1,
    mode='min',
)
```

```python
model.fit(x=x_train, y=y_train, epochs = 20, validation_data = (x_test,y_test), verbose = 1, callbacks=[early_Stop])
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
deneme=model.predict(x_test)
```

```python
tahmin = (deneme > 0.5).astype("int32").flatten()
tahmin
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahmin))
```

```python
print(confusion_matrix(y_test,tahmin))
```
